```python
# Importando as bibliotecas e comando %matplotlib:
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# Carregando arquivo no notebook e fazendo a leitura:
def_amazon_data = pd.read_csv('/content/def_area_2004_2019.csv')
```

```python
# Exibindo o cabe√ßalho do Dataframe:
def_amazon_data.head()
```

```python
# Renomeando a primeira e a √∫ltima coluna para melhor entendimento:
def_amazon_data = def_amazon_data.rename(columns={'Ano/Estados': 'Ano', 'AMZ LEGAL': 'Total'})
```

```python
# Exibindo o cabe√ßalho do Dataframe ap√≥s as 2 colunas serem renomeadas:
def_amazon_data.head()
```

```python
# Foi verificado que o Dataset n√£o apresenta valores nulos ao utilizar o comando: 
def_amazon_data.isnull().sum()
```

```python
# plotando as √°reas totais devastadas atrav√©s dos anos:
plt.figure(figsize=(12,8))
plt.suptitle('√Årea total desmatada por ano na Amaz√¥nia', fontsize=14, weight='bold')
plt.subplot(2,1,1)
plt.bar(def_amazon_data.Ano, def_amazon_data.Total, color='b', alpha=0.6)
plt.ylabel('Total',fontsize=14)
plt.xticks(def_amazon_data.Ano)

plt.subplot(2,1,2)
plt.plot(def_amazon_data.Ano, def_amazon_data.Total)
plt.xlabel('Ano', fontsize=14)
plt.ylabel('Total',fontsize=14)
plt.xticks(def_amazon_data.Ano)
fig = plt.gcf()
plt.savefig('grafico1.png', format='png')
plt.show()
```

```python
# Plotando a soma da √°rea desmatada total para cada estado e para todos os anos no conjunto de dados:
states=["AC","AM","AP","MA","MT","PA","RO","RR","TO"]
data=[]
for i in range(len(states)):
    data.append(def_amazon_data[states[i]].sum())
plt.figure(figsize=(12,4))
plt.suptitle('√Årea total desmatada por estado', fontsize=14, weight='bold')
plt.bar(states, data, color='darkolivegreen',alpha=1.0)
plt.xlabel('Estados', fontsize=14)
plt.ylabel('Total',fontsize=14)
fig = plt.gcf()
plt.savefig('grafico2.png', format='png')
plt.show()  
```

```python
# Plotando a √°rea desmatada anualmente por estado de 2004 a 2019:
states=["AC","AM","AP","MA","MT","PA","RO","RR","TO"]
fig, ax = plt.subplots(3, 3, figsize=(15,10))
plt.suptitle('√Årea amaz√¥nica desmatada anualmente por estado - 2004 a 2019', fontsize=15, weight='bold')
plt.subplots_adjust(wspace=0.3, hspace=0.5)
col_names = def_amazon_data.columns[1:10]
for i in range(3):
    for j in range(3):
        col = col_names[i*3+j]
        ax[i,j].plot(def_amazon_data['Ano'], def_amazon_data[col], linewidth=2, color='forestgreen')
        ax[i,j].set_ylabel('√Årea em $km^2$', fontsize=12)
        ax[i,j].set_title(states[i*3+j])
        ax[i,j].grid(axis='y', color='0.95')
        ax[i,j].xaxis.set_major_locator(plt.MaxNLocator(integer=True))
        if (j>0):
             ax[i,j].set(ylabel='')
fig = plt.gcf()
plt.savefig('grafico3.png', format='png')

```
